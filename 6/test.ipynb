```python
from rapidfuzz import process, fuzz
```

```python
import pandas as pd
```

```python
df = pd.read_parquet('employees.parquet')
```

```python
orcs = pd.read_parquet('orcs.parquet')
```

```python
from joblib import Parallel, delayed
```

```python
def get_fingerprint(df):
    df.fillna('<nan>', inplace=True)
    return df['name'] + ' ' + df['surname'] + ' ' + df['fathername'] + ' ' + df['birthdate'] + ' ' + df['inn'] + ' ' + df['passport'] 
```

```python
df1 = get_fingerprint(df)
```

```python
df2 = get_fingerprint(orcs)
```

```python
from tqdm_joblib import tqdm_joblib
```

```python
with tqdm_joblib("Processing", total=len(df2)):
    candidates = Parallel(n_jobs=-1, batch_size=100)(delayed(lambda row: process.extract(row,
                                                                                         df1, scorer=fuzz.WRatio, limit=5))(row) for row in df2)
```
