```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./active_learning_data/valid.csv')
```

```python
df = df[df.abs() < 10]
```

```python
for feature in df.drop(columns='y').columns:
    plt.scatter(df[feature], df['y'])
    plt.show()
```

```python
df[df['y'].isna()]
```

```python
df.dropna(axis=0, inplace=True)
```

```python
X = df.drop(columns='y')
y = df['y']
```

```python
from sklearn.decomposition import PCA, TruncatedSVD
```

```python
mapper = PCA(n_components=5).fit(X, y)
```

```python
X_tr = mapper.transform(X)
```

```python
for i in range (X_tr.shape[1]):
    plt.scatter(X_tr[:, i], y)
    plt.show()
```
